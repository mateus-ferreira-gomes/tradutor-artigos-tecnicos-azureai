# 📘 Tradução de Artigos Técnicos da Web com Azure OpenAI

Este notebook demonstra como:

- Extrair conteúdo textual de páginas web
- Processar o conteúdo removendo elementos HTML desnecessários
- Traduzir o texto utilizando o Azure OpenAI

Fluxo do notebook:

URL do artigo → Extração do texto → Tradução com Azure OpenAI → Saída em Markdown

```python
# 📌 Importação das bibliotecas

# Importa a biblioteca requests para fazer requisições HTTP
import requests

# Importa a classe BeautifulSoup da biblioteca bs4
# Usada para analisar o HTML das páginas
from bs4 import BeautifulSoup
```

```python
# 📌 Função para extrair texto de uma URL

# Importa a biblioteca requests para fazer requisições HTTP (acessar páginas da internet)
import requests

# Importa a classe BeautifulSoup da biblioteca bs4 para analisar o HTML das páginas
from bs4 import BeautifulSoup


# Define uma função chamada extract_text_from_url que recebe uma URL como parâmetro
def extract_text_from_url(url):

    # Faz uma requisição HTTP GET para acessar o conteúdo da página informada na URL
    response = requests.get(url)

    # Verifica se a requisição foi bem sucedida (código 200 significa sucesso)
    if response.status_code == 200:

        # Cria um objeto BeautifulSoup para interpretar o HTML retornado pela página
        # 'html.parser' é o interpretador utilizado para ler o HTML
        soup = BeautifulSoup(response.text, 'html.parser')

        # Procura todas as tags <script> e <style> dentro do HTML
        for script_or_style in soup(['script', 'style']):

            # Remove essas tags do documento, pois não fazem parte do conteúdo principal
            script_or_style.decompose()

    else:
        # Caso a requisição falhe, mostra uma mensagem com o código de erro retornado
        print(f"Failed to fetch the URL. Status code: {response.status_code}")

        # Retorna None indicando que não foi possível obter o conteúdo da página
        return None

    # Extrai todo o texto da página HTML já processada
    text = soup.get_text()

    # Retorna o texto extraído da página
    return text


# Chama a função passando a URL de um artigo técnico
extract_text_from_url('https://dev.to/kenakamu/azure-open-ai-in-vnet-3alo')
```

```python
# 📌 Importação da biblioteca Azure OpenAI

# Importa o modelo de chat do Azure OpenAI através da biblioteca LangChain
from langchain_openai.chat_models.azure import AzureChatOpenAI
```

```python
# 📌 Configuração do cliente Azure OpenAI

# Cria o cliente que fará conexão com o serviço Azure OpenAI
client = AzureChatOpenAI(

    # Endpoint da instância criada no Azure OpenAI
    azure_endpoint="https://openai-mateus.openai.azure.com/",

    # Chave de API usada para autenticar o acesso ao serviço
    # ⚠️ Substitua o valor abaixo pela sua chave real no seu ambiente local
    # Não deixe a chave real aqui no GitHub!
    api_key="YOUR_AZURE_OPENAI_KEY",

    # Versão da API do Azure OpenAI que será utilizada
    api_version="2024-02-15-preview",

    # Nome do deployment criado no Azure (modelo disponibilizado)
    deployment_name="gpt-4o-mini",

    # Número máximo de tentativas caso ocorra erro na requisição
    max_retries=0
)
```

```python
#  📌 Função de tradução de artigos

# Define uma função chamada translate_article
# Ela recebe dois parâmetros:
# text → texto que será traduzido
# lang → idioma para o qual o texto será traduzido
def translate_article(text, lang):

    # Cria uma lista de mensagens que serão enviadas para o modelo
    messages = [

        # Mensagem de sistema que define o comportamento da IA
        ("system", "Você atua como tradutor de textos"),

        # Mensagem do usuário contendo a instrução de tradução
        ("user", f"Traduza o {text} para o idioma {lang} e responda em markdown")
    ]

    # Envia as mensagens para o modelo Azure OpenAI
    response = client.invoke(messages)

    # Imprime o conteúdo da resposta gerada pelo modelo
    print(response.content)

    # Retorna o texto traduzido
    return response.content
```

```python
# 📌 Teste da função de tradução

# Chama a função translate_article
# Envia uma frase em inglês para ser traduzida
# Define o idioma de saída como português
translate_article("Let's see if the deployment was succeeded.", "portuguese")
```

```python
# 📌 Tradução de artigo a partir de uma URL

url = 'https://dev.to/kenakamu/azure-open-ai-in-vnet-3alo'
text = extract_text_from_url(url)
article = translate_article(text, "pt-br")

print(article)
```
